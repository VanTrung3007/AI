```python
import pandas as pd
import matplotlib.pyplot as plt
url = 'winequalityN.csv'
data = pd.read_csv(url)
data.hist()
plt.show
```

```python
import seaborn as sns
correlationdata = data.corr()
sns.heatmap(correlationdata, annot = True)
```

```python
data.plot(kind = 'scatter', x = 'free sulfur dioxide', y = 'total sulfur dioxide', figsize=(9,9), color = 'black')
```

```python
#Mô hình bậc 1
from sklearn import linear_model
modeldata = linear_model.LinearRegression()
X = pd.DataFrame(data['free sulfur dioxide'])
modeldata.fit(X, y = data['total sulfur dioxide'])
print('Intercept của mô hình: ', modeldata.intercept_)
print('Coeffient của mô hình: ', modeldata.coef_)
print('Độ chính xác của mô hình: ', 
      modeldata.score(X,y = data['total sulfur dioxide']))
data.plot(kind = 'scatter', x = 'free sulfur dioxide', y = 'total sulfur dioxide', figsize=(9,9), 
                 color = 'black')
y_pred = modeldata.predict(X)
plt.plot(X,y_pred, color = 'blue')
```

```python
#Mô hình bậc 10
from sklearn import linear_model
modeldata = linear_model.LinearRegression()
X = pd.DataFrame([data['free sulfur dioxide'], data['free sulfur dioxide']**2, data['free sulfur dioxide']**3, data['free sulfur dioxide']**4,
                  data['free sulfur dioxide']**5, data['free sulfur dioxide']**6,data['free sulfur dioxide']**7, data['free sulfur dioxide']**8,
                  data['free sulfur dioxide']**9, data['free sulfur dioxide']**10]).T
modeldata.fit(X, y = data['total sulfur dioxide'])
print('Intercept của mô hình: ', modeldata.intercept_)
print('Coeffient của mô hình: ', modeldata.coef_)
print('Độ chính xác của mô hình: ', 
      modeldata.score(X,y = data['total sulfur dioxide']))
data.plot(kind = 'scatter', x = 'free sulfur dioxide', y = 'total sulfur dioxide', figsize=(9,9), 
                 color = 'black')
y_pred = modeldata.predict(X)
plt.plot(X,y_pred, color = 'blue')
```
