```python
import pandas as pd
import matplotlib.pyplot as plt
url = 'lung_cancer_examples.csv'
data = pd.read_csv(url)
data.hist()
plt.show
```

```python
correlationdata = data.corr()
import seaborn as sns
sns.heatmap(correlationdata, annot = True)
```

```python
data.plot(kind = 'scatter', x = 'Alkhol', y = 'Result', figsize=(9,9), color = 'black')
```

```python
#Mô hình bậc 1
from sklearn import linear_model
modeldata = linear_model.LinearRegression()
X = pd.DataFrame(data['Alkhol'])
modeldata.fit(X, y = data['Result'])
print('Intercept của mô hình: ', modeldata.intercept_)
print('Coeffient của mô hình: ', modeldata.coef_)
print('Độ chính xác của mô hình: ', 
      modeldata.score(X,y = data['Result']))
data.plot(kind = 'scatter', x = 'Alkhol', y = 'Result', figsize=(9,9), 
                 color = 'black')
y_pred = modeldata.predict(X)
plt.plot(X,y_pred, color = 'blue')
     
```

```python
#Mô hình bậc 10
from sklearn import linear_model
modelHousing10 = linear_model.LinearRegression()
X = pd.DataFrame([data['Alkhol'], data['Alkhol']**2, data['Alkhol']**3, data['Alkhol']**4,
                  data['Alkhol']**5, data['Alkhol']**6,data['Alkhol']**7, data['Alkhol']**8,
                  data['Alkhol']**9, data['Alkhol']**10]).T
modelHousing10.fit(X, y = data['Result'])
print('Intercept của mô hình: ', modelHousing10.intercept_)
print('Coeffient của mô hình: ', modelHousing10.coef_)
print('Độ chính xác của mô hình: ', 
      modelHousing10.score(X,y = data['Result']))

import numpy as np
range = np.arange(0,8,0.1)
data.plot(kind = 'scatter', x = 'Alkhol', y = 'Result', figsize=(9,9),
                 color = 'black', xlim=(0,8))
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5,
                          range**6, range**7, range**8, range**9, range**10]).T
y_pred = modelHousing10.predict(X = predictor)
plt.plot(range,y_pred, color = 'blue')
```
