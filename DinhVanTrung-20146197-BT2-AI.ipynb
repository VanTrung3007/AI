Lọc huy chương của 1 nước và sắp xếp theo các năm

```python
import pandas as pd       
URL = 'http://winterolympicsmedals.com/medals.csv'
dataframe = pd.read_csv(URL)
```

```python
GER = dataframe[dataframe['NOC']=='GER']    #Lọc nước Đức
GER.head(200)
```

```python
GER['Sport'].value_counts()       #Số môn thể thao nước Đức đã giành được huy chương
```

```python
x = GER.drop(GER.columns[0:7],axis=1)  # Tổng số huy chương nước Đức nhận được từ 1928 tới 2006
x.count()
```

```python
x.value_counts()      # Tổng số huy chương nhận được phân loại theo bạc, đồng, vàng của nước Đức
```

```python
y = GER.drop(['NOC','Event','City','Discipline','Event gender'],axis=1)   # Số huy chương theo các năm của nước Đức
y.groupby(["Year","Medal","Sport"]).count()
```

```python
y.groupby(["Sport","Medal"]).count()      #Số huy chương nhận được theo môn thể thao
```
